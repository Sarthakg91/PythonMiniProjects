```python
# %load LIWCMeta.py

"""
Created on Fri Mar 01 13:35:34 2016

@author: Sarthak Ghosh
"""

print positiveListCopy[0:5]

import re

def getLex(post, liwc_lexicons):
	liwc_lexicon_to_count = {}
	for liwc_lexicon_name in set(liwc_lexicons.keys()):
		liwc_lexicon_to_count[liwc_lexicon_name] = 0	
	for name, items in liwc_lexicons.iteritems():
		for item in items:
			pattern = re.compile(item)
			count = len(pattern.findall(post))
			liwc_lexicon_to_count[name] += count
		liwc_lexicon_to_count[name] /= float(len(post.split(' ')))
# 		if liwc_lexicon_to_count[name]< 0.0:
# 			print "negative :", post
# 		if name=="positive_affect" and liwc_lexicon_to_count[name] == 1.0:
# 			print "hi: ", post
	return liwc_lexicon_to_count

def get_liwc_dictionary():
	
	liwc_lex_dict = {}
	liwc_lex_dict["positive_affect"] = get_liwc_lexicons("positive_affect")
	liwc_lex_dict["negative_affect"] = get_liwc_lexicons("negative_affect")
	liwc_lex_dict["anger"] = get_liwc_lexicons("anger")
	liwc_lex_dict["anxiety"] = get_liwc_lexicons("anxiety")	
	liwc_lex_dict["sadness"] = get_liwc_lexicons("sadness")
	liwc_lex_dict["swear"] = get_liwc_lexicons("swear")
	return liwc_lex_dict



def get_liwc_lexicons(category):
	liwc_lexicons_directory = "LIWC_lexicons"
	lexicon = []
	lexicon_str = "("
	with open("{0}/{1}".format(liwc_lexicons_directory, category), "r") as file_handle:
		for line in file_handle:
			item = line.strip()
			if "*" in item:
				item = r"\b{0}\b".format(item.replace("*", ".*?"))
			else:
				item = r"\b{0}\b".format(item)
			lexicon_str += item + "|"
	lexicon_str = lexicon_str[:-1] + ")"
	lexicon.append(lexicon_str)
	return lexicon

liwc_dictionary=get_liwc_dictionary()
#print liwc_dictionary


with open("negativeVectors.tsv", "w") as negativeVectors_file:
    for i in range(len(negativeListCopy)):
        outputDictionaryNegative=getLex(negativeListCopy[i], liwc_dictionary)
        negativeVectors_file.write(str(negativeListCopy[i])+"\t"+str(outputDictionaryNegative["negative_affect"])+"\t"+str(outputDictionaryNegative["positive_affect"])+"\t"+str(outputDictionaryNegative["anger"])
                                   +"\t"+str(outputDictionaryNegative["anxiety"])+"\t"+str(outputDictionaryNegative["swear"])+"\t"+str(outputDictionaryNegative["sadness"])+"\n")    

with open("positiveVectors.tsv", "w") as positiveVectors_file:
    for i in range(len(positiveListCopy)):
        outputDictionaryPositive=getLex(positiveListCopy[i], liwc_dictionary)
        positiveVectors_file.write(str(positiveListCopy[i])+"\t"+str(outputDictionaryPositive["negative_affect"])+"\t"+str(outputDictionaryPositive["positive_affect"])+"\t"+str(outputDictionaryPositive["anger"])
                                   +"\t"+str(outputDictionaryPositive["anxiety"])+"\t"+str(outputDictionaryPositive["swear"])+"\t"+str(outputDictionaryPositive["sadness"])+"\n")    


        
print negativeListCopy[0:5]
print positiveListCopy[0:5]
    









```

```python
#part b
#removing stopwords from tweets
import nltk
from nltk.corpus import stopwords

print positiveListCopy[0:5]

def removeStopWords(sentence):
    flitered_word_list=[]
    word_list=sentence.split(' ')
    filtered_word_list = [word for word in word_list if word not in stopwords.words('english')]
    return (' '.join(filtered_word_list)).strip()

filteredNegativeList=[]
filteredPositiveList=[]
for post in negativeListCopy:
#     print post
    filteredNegativeList.append(removeStopWords(post))
for post in positiveListCopy:
#     print post
    filteredPositiveList.append(removeStopWords(post))
    
    
print filteredPositiveList[0:5]
print filteredNegativeList[0:5]

allPostsList=filteredNegativeList+filteredPositiveList

print len(allPostsList)

# print len(filteredNegativeList)
# print len(filteredPositiveList)
# print len(allPostsList)
# print filteredPositiveList[-1]




```

```python
#calculating n-grams for all posts
from nltk.util import ngrams
from collections import Counter


listOfUnigrams=[]
listOfBigrams=[]
listOfTrigrams=[]
unigrams=[]
bigrams=[]
trigrams=[]


for post in allPostsList: 
    wordList=post.split(' ')
    unigrams=ngrams(wordList,1)
    bigrams=ngrams(wordList,2)
    if (len(wordList)>3):
        trigrams=ngrams(wordList,3)        
    for unigram in unigrams:
        listOfUnigrams.append(unigram)
    for bigram in bigrams:
        listOfBigrams.append(bigram)
    for trigram in trigrams:
        listOfTrigrams.append(trigram)
        
# print "lenght of trigrams : "+ str(len(listOfTrigrams))
# print "first Trigram is : "+str(listOfTrigrams[0])

fdistUnigrams = nltk.FreqDist(listOfUnigrams)
fdistBigrams =  nltk.FreqDist(listOfBigrams)
fdistTrigrams = nltk.FreqDist(listOfTrigrams)

funi=[]
fbi=[]
ftri=[]

#sort according to the frequencies
for k,v in fdistUnigrams.items():
    funi.append((k,v))
funi.sort(key=lambda tup: tup[1], reverse=True)

for k,v in fdistBigrams.items():
    fbi.append((k,v))
fbi.sort(key=lambda tup: tup[1], reverse=True)

for k,v in fdistTrigrams.items():
    ftri.append((k,v))
ftri.sort(key=lambda tup: tup[1], reverse=True)

#combine all lists of frquency distributions of ngrams
fn=funi+fbi+ftri
print len(fn)

fn.sort(key=lambda tup:tup[1], reverse=True)


# print type(fn)
fn500=[]
fn100=[]
fn50=[]

for k in fn:
    if k[1]>=500:
        fn500.append(k)
    if k[1]>=100:
        fn100.append(k)
    if k[1]>50:
        fn50.append(k)

#here we have list of all ngrams that have frequencies above 50, 100 and 500            

print len(fn500), len(fn50), len(fn100)
```

```python
# code for KNN classifer.
#there should be 4 KNN classifers with k=1,10,100 and 1000
from sklearn.cross_validation import StratifiedKFold
import numpy as np
import numpy as np

from sklearn.neighbors import KNeighborsClassifier
def fitKNN(x_train,x_test,y_train,y_test,knn=1,):
    neigh = KNeighborsClassifier(n_neighbors=knn)    
    y_pred = neigh.fit(x_train, y_train).predict(x_test)
    return [accuracy_score(y_test,y_pred), f1_score(y_test,y_pred),recall_score(y_test,y_pred),precision_score(y_test,y_pred)]
    
    
    

#code for Naive Bayes Classifier
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score,f1_score,recall_score,precision_score
def fitNaiveBayes(x_train,x_test,y_train,y_test):
    clf_NB = GaussianNB()
    y_pred = clf_NB.fit(x_train, y_train).predict(x_test)
    return [accuracy_score(y_test,y_pred), f1_score(y_test,y_pred),recall_score(y_test,y_pred),precision_score(y_test,y_pred)]
#     GaussianNB()
#     now you can use clf.predict(dataRow)


#classifier code for SVM
from sklearn.svm import SVC
def fitSVM(x_train,x_test,y_train,y_test):    
    clf_SVM = SVC()
    y_pred = clf_SVM.fit(x_train, y_train).predict(x_test)
    return [accuracy_score(y_test,y_pred), f1_score(y_test,y_pred),recall_score(y_test,y_pred),precision_score(y_test,y_pred)]
    

    
    


```

```python
def loadDataForClassification(fileNamePositive,fileNameNegative,n):
    nFeatureVectors=[]
    temp=[]
    X=[]
    Y=[]
    nFeatureVectors=open(fileNameNegative).read().splitlines()
    for vector in nFeatureVectors[0:n]:
        temp=vector.split('\t')
        temp=temp[1:]
        X.append([float(i) for i in temp])
        Y.append(0)
    temp[:]=[]
    pFeatureVectors=open(fileNamePositive).read().splitlines()
    for vector in pFeatureVectors[0:n]:
        temp=vector.split('\t')
        temp=temp[1:]
        X.append([float(i) for i in temp])
        Y.append(1)
    return [X,Y]

dataset=loadDataForClassification("positiveVectors.tsv","negativeVectors.tsv",1000)
X=dataset[0]
Y=dataset[1]
print X
```

```python
def calculateAllScores(X,Y):
    X, Y= np.array(X), np.array(Y)    
    k_fold_accuracies_NB = []
    k_fold_accuracies_SVM = []
    k_fold_accuracies_KNN1 = []
    k_fold_accuracies_KNN10 = []
    k_fold_accuracies_KNN100 = []
    k_fold_accuracies_KNN1000 = []
    scores_NB=[]
    scores_SVM=[]
    scores_KNN1=[]
    scores_KNN10=[]
    scores_KNN100=[]
    scores_KNN1000=[]
    skf = StratifiedKFold(Y, n_folds=5)
    for train_index, test_index in skf:
        k_fold_accuracies_NB.append(fitNaiveBayes(X[train_index],X[test_index],Y[train_index],Y[test_index]))
        k_fold_accuracies_SVM.append(fitSVM(X[train_index],X[test_index],Y[train_index],Y[test_index]))
        k_fold_accuracies_KNN1.append(fitKNN(X[train_index],X[test_index],Y[train_index],Y[test_index],1))
        k_fold_accuracies_KNN10.append(fitKNN(X[train_index],X[test_index],Y[train_index],Y[test_index],10))
        k_fold_accuracies_KNN100.append(fitKNN(X[train_index],X[test_index],Y[train_index],Y[test_index],100))
        k_fold_accuracies_KNN1000.append(fitKNN(X[train_index],X[test_index],Y[train_index],Y[test_index],1000))
    scores_NB=np.asarray(k_fold_accuracies_NB).mean(axis=0)
    scores_SVM=np.asarray(k_fold_accuracies_SVM).mean(axis=0)
    scores_KNN1=np.asarray(k_fold_accuracies_KNN1).mean(axis=0)
    scores_KNN10=np.asarray(k_fold_accuracies_KNN10).mean(axis=0)
    scores_KNN100=np.asarray(k_fold_accuracies_KNN100).mean(axis=0)
    scores_KNN1000=np.asarray(k_fold_accuracies_KNN1000).mean(axis=0)

    return [scores_NB,scores_KNN1,scores_SVM,scores_KNN10,scores_KNN100,scores_KNN1000]
```

### b. Why do you think some classifiers perform better than the others?
 We have used KNN, SVM and Naive Bayes classifiers and from the tables we obtained, we can see that certain classifiers perform better than others. This is true for all datasets and not just the current one in use. The performance of classifiers depends on a number of things including the type of the data set, the number of data points and values of the tuning parameters. While KNN is better for data points that show clustering behaviour, but it needs to be carefully tuned. Depending on the value of K, accuracies can vary. It also doesn't perform well when there are many data points. SVM is a very popular algorithm which works better with a limited set of points in a high dimensional space. SVM is good with outliers as it will only use the most relevant points to find a linear separation (support vectors). Naive Bayes on the other hand, assumes that the data points are conditionally independent. If that is not the case and there are some hidden correlated attributes, then the performance of NAive Bayes algorithm will be greatly hampered. 
 

### c. How do the different feature choices impact classifier performance? 

Features are like decision points. Every data-point is represented using a set of features. These features in turn help the machine decide which data point it is. For some data classes, certain types of features are uniquely associated and must be included for good performance of a classifier. For example, if you are classifying between dogs and cats and you do not include the size of the animal, or the ear-length of the animal as feature (dogs are usually larger than cats and they have longer ears), then the classifer performance will certainly go down. In the paper, "Impact of Feature Selection Techniques for Tweet Sentiment Classification", Prusa et. al have looked this feature selection problem specifically for sentiment analyses of tweets. This paper has evaluated the performance of ten filter-based feature ranking techniques on tweets labelled as positive and negative.There results show that sentiment classification can indeed be greatly affected by feature selection. 
They report that, " In particular five feature ranking techniques (Chi-Squared, the Kolmogorov-Smirnov statistic, Mutual Information, area under the Precision-Recall curve, and area under the Receiver Operating Characteristic curve) improve classification performance over no feature selection". Moreover the authors also show that using 75-200 features can improve classification results on a whole.
